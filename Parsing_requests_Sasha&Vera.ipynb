```python
import requests
import numpy as np
import pandas as pd
```

```python
url = 'https://stockscan.io/stocks/AAPL/price-history' 
page = requests.get(url)
page
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
print(soup.prettify())
```

```python
quotes = []
first_table = soup.find('table')
for row in first_table.find_all('tr'):
    quotes.append(row.text.split('\n')[1:-1])
print(quotes)
```

```python
headers = quotes[0]
table = quotes[1:]
df = pd.DataFrame(table, columns=headers)
df
```
