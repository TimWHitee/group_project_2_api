```python
import requests
import numpy as np
import pandas as pd
```

Для начала мы рассмотрели сайт stockscan.io. В качестве теста смотрели по компании Apple 

```python
url = 'https://stockscan.io/stocks/AAPL/price-history' 
page = requests.get(url)
page
```

Получив от него response 200, мы продолжили работу

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(page.text, 'html')
soup
```

Чтобы лучше проанализировать код до начала работы с ним, мы использовали soup.prettify():

```python
print(soup.prettify())
```

```python
import logging
from logger_config import logs, details

if logs and details:
    logging.basicConfig(level=logging.INFO, filename="py_log.log",filemode="a",
                    format="%(asctime)s %(levelname)s %(message)s")
    try:
        quotes = []
        first_table = soup.find('table')
        for row in first_table.find_all('tr'):
            quotes.append(row.text.split('\n')[1:-1])
            logging.info("file: Parsing_requests_Sasha&Vera.ipynb. Row added")
        print(quotes)
        logging.info("file: Parsing_requests_Sasha&Vera.ipynb. Success")
    except Exception as e:
        logging.error(f"file: Parsing_requests_Sasha&Vera.ipynb. Error {e}")
        
elif logs and (details == False):
    logging.basicConfig(level=logging.INFO, filename="py_log.log",filemode="a",
                    format="%(asctime)s %(levelname)s %(message)s")
    try:
        quotes = []
        first_table = soup.find('table')
        for row in first_table.find_all('tr'):
            quotes.append(row.text.split('\n')[1:-1])
        print(quotes)
        logging.info("file: Parsing_requests_Sasha&Vera.ipynb. Success")
    except Exception as e:
        logging.error(f"file: Parsing_requests_Sasha&Vera.ipynb. Error {e}")

else:
    quotes = []
    first_table = soup.find('table')
    for row in first_table.find_all('tr'):
        quotes.append(row.text.split('\n')[1:-1])
    print(quotes)
```

После, из данных, полученных парсингом, был получен pd.dataframe:

```python
headers = quotes[0]
table = quotes[1:]
df = pd.DataFrame(table, columns=headers)

df
```
